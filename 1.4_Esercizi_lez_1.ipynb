### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
# Ciclo 1:
dispari_1 = []
for n in range(9):
    if n%2 != 0:
        dispari_1.append(n)
print("Dispari_1: ", dispari_1)

dispari_2 = [n for n in range(9) if n%2 != 0]
print("Dispari_2: ", dispari_2)

# Ciclo 2:
Input = [[1,2,3], [4,5], [6,7,8,1]]
Output_1 = []
for lista in Input:
   for i in lista:
      Output_1.append(i)
print("Output_1: ", Output_1)

Output_2 = [i for lista in Input for i in lista]
print("Output_2: ", Output_2)

# Ciclo 3:
lista_a = [0,1,3,4]
lista_b = ['a', 'b', 'c']

lista_new_1 = []
for x in lista_a:
    for y, elem in enumerate(lista_b):
        if x%2 == 0 and y%2 != 0:
            lista_new_1.append((x, elem))
print("Lista_new_1: ", lista_new_1)            

lista_new_2 = [(x, elem) for x in lista_a for y, elem in enumerate(lista_b) if x%2 == 0 and y%2 != 0]
print("Lista_new_2: ", lista_new_2)
```

### **2. Vettore di Numeri Primi**

1. **Crea un vettore** contenente tutti i numeri primi compresi tra **0 e 10**  
   _(Puoi scriverli direttamente nell'array)._

2. **Conta quanti numeri ci sono nel vettore** utilizzando la funzione `len()`.  **Ottieni lo stesso numero** accedendo all'attributo `.size` del vettore.

3. Quale pensi sia il tipo di dato (`dtype`) del vettore?  Prova a rispondere senza eseguire il codice e Verifica la tua risposta accedendo all'attributo `.dtype` del vettore

4. Scrivi l'array usando una list comprehension che controlla che i numeri siano primi. 

```python
import numpy as np

# Step 1:
v = (1,2,3,5,7)
a = np.array([1,2,3,5,7])     # array contenente tutti i numeri primi tra 0 e 10
print("Numeri primi: ", a)

# Step 2:
l = len(a)
print("Lung a: ", l)

s = a.size
print("Lung a: ", s)

# Step 3: il tipo di dato dovrebbe essere int
t_a = a.dtype
print("Tipo di dato: ", t_a)

# Step 4:
def is_primo(x):
    for i in range(2, x):
        if x%i == 0:
            return False
    return True

primi = np.array([i for i in range(2,10) if is_primo(i)])
print("Numeri primi: ", primi)
```

### **3.  Operazioni su array**
Crea un array **a**, 1D (senza digitarlo esplicitamente), e genera due nuovi array **b**, contenente una sottostringa a piacere e **c** che è il reverse dell'array **a** (esempio [1,2,3] diventa [3,2,1]) .  Dividi l'array **a** per l'array **c**. Fai la stessa cosa per una lista.

```python
# Con array:
a = np.arange(1, 20)
b = a[2:4]
c = a[::-1]
d = a/c

print("Array a: ", a)
print("Array b: ", b)
print("Array c: ", c)
print("Array d: ", d)

# Con lista:
a = list(range(1, 20))
b = a[2:4]
c = a[::-1]
d = []
for i in range(len(a)):
    d.append(a[i]/c[i])

d_comp = [a[i] / c[i] for i in range(len(a))]

print("Lista a: ", a)
print("Lista b: ", b)
print("Lista c: ", c)
print("Lista d: ", d)
print("Lista d con lista comprehention: ", d_comp)
```
